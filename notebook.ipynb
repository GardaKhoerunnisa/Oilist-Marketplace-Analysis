### **Tabel Customers**

### **Tabel Orders**

```python
order=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/olist_orders_dataset.csv')
order.head()
```

### **Tabel Order Item**

```python
order_item=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/olist_order_items_dataset.csv')
order_item.head()
```

### **Tabel Product**

```python
products=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/olist_products_dataset.csv')
products.head()
```

### **Tabel Seller**

```python
sellers=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/olist_sellers_dataset.csv')
sellers.head()
```

### **Tabel Order Payments**

```python
order_payments=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/olist_order_payments_dataset.csv')
order_payments.head()
```

### **Tabel Category Translation**

```python
category_translation=pd.read_csv('/content/drive/MyDrive/Proyek Analisis Data/product_category_name_translation.csv')
category_translation.head()
```

## **Menggabungkan Data**

```python
# Menggabungkan orders dengan order_items
order_data = pd.merge(order, order_item, on='order_id', how='inner')

# Menggabungkan order_data dengan products
order_data = pd.merge(order_data, products, on='product_id', how='inner')

# Menggabungkan order_data dengan customers
order_data = pd.merge(order_data, customers, on='customer_id', how='inner')

# Menggabungkan order_data dengan order_payments
order_data = pd.merge(order_data, order_payments, on='order_id', how='inner')

# Menggabungkan order_data dengan sellers
order_data = pd.merge(order_data, sellers, on='seller_id', how='inner')

# Menggabungkan dengan category_translation untuk mendapatkan nama kategori produk yang lebih jelas
order_data = pd.merge(order_data, category_translation, on='product_category_name', how='left')
```

##**Assessing Data & Exploratory Data**

```python
print(order_data.isnull().sum())
```

```python
#Check Data Duplicate
print("Jumlah duplikasi: ", order_data.duplicated().sum())
```

## **Cleaning Data**

```python
#menangani kesalahan tipe data
datetime_columns = ["order_purchase_timestamp", "order_approved_at",
                    "order_delivered_carrier_date", "order_delivered_customer_date",
                    "order_estimated_delivery_date","shipping_limit_date"]

for column in datetime_columns:
    order_data[column] = pd.to_datetime(order_data[column], errors='coerce')
```

```python
order_data.info()
```

```python
#menangani missing value
order_data.isna().sum()
```

```python
#menangani missing value
order_data["order_purchase_timestamp"].fillna(order_data['order_status'], inplace=True)
order_data["order_estimated_delivery_date"].fillna(order_data['order_status'], inplace=True)
order_data["shipping_limit_date"].fillna(order_data['order_status'], inplace=True)
order_data["order_approved_at"].fillna(order_data['order_status'], inplace=True)
order_data["order_delivered_customer_date"].fillna(order_data['order_status'], inplace=True)
order_data["order_delivered_carrier_date"].fillna(order_data['order_status'], inplace=True)
```

```python
order_data.isna().sum()
```

```python
#menangani missing value
order_data["product_category_name"].fillna("Unknown", inplace=True)

columns_to_drop = [
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty',
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm'
]
order_data = order_data.drop(columns=columns_to_drop)
order_data.isna().sum()
```

```python
order_data["product_category_name_english"].fillna('Others', inplace=True)
```

```python
order_data.isna().sum()
```

```python
order_data.info()
```

##**Menyimpan Data yang Bersih**

```python
order_data.to_csv('cleaned_order_data.csv', index=False)
```

## **Distribusi Penjualan Berdasarkan Kategori Produk**

```python
top_categories = order_data['product_category_name_english'].value_counts().head(10).reset_index()
top_categories.columns = ['product_category_name_english', 'sales_count']

print("Tabel 10 Kategori Produk Terlaris:")
print(top_categories)
```

```python
plt.figure(figsize=(14,7))
cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(0, 1, len(top_categories)))

plt.barh(top_categories['product_category_name_english'], top_categories['sales_count'], color=colors)
plt.xlabel('Jumlah Penjualan')
plt.ylabel('Kategori Produk')
plt.title('10 Kategori Produk Terlaris')
plt.gca().invert_yaxis()
plt.show()
```

## **Distribusi Penjualan Berdasarkan Negara Pelanggan**

```python
state_sales = order_data['customer_state'].value_counts().reset_index()
state_sales.columns = ['customer_state', 'sales_count']

# Menampilkan tabel data
print("Tabel Distribusi Penjualan Berdasarkan Negara Bagian Pelanggan:")
print(state_sales)
```

```python
plt.figure(figsize=(14,7))
cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(0, 1, len(state_sales)))

plt.bar(state_sales['customer_state'], state_sales['sales_count'], color=colors)
plt.xlabel('Negara Bagian')
plt.ylabel('Jumlah Penjualan')
plt.title('Distribusi Penjualan Berdasarkan Negara Pelanggan')
plt.xticks(rotation=90)
plt.show()
```

## **Distribusi Metode Pembayaran**

```python
payment_method_distribution = order_data['payment_type'].value_counts().reset_index()
payment_method_distribution.columns = ['payment_type', 'count']

# Menampilkan tabel data
print("Tabel Distribusi Metode Pembayaran:")
print(payment_method_distribution)

```

```python
plt.figure(figsize=(14,7))
plt.bar(payment_method_distribution['payment_type'], payment_method_distribution['count'], color='skyblue')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Jumlah Pembayaran')
plt.title('Distribusi Metode Pembayaran')
plt.show()
```
